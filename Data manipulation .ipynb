# Sorting and Subsetting 

## Sorting

### 1) Sorting by single variables 

```python
# sorting using sort_values() function
# this sorts data in accordance to weight in a desending order
df.sort_values('weight_kg')
```

```python
# sorting in asecending order
# this will sort in accordance of height in ascending order
df.sort_values('height_cm', ascending= False)
```

### 2) Sorting by multiple variables 

```python
# use sort_values() function
df.sort_values(['height_cm','weight_kg'], ascending=[False,True ])
```

## loc

#### The .loc[ ] function in Pandas is used for label-based indexing, allowing you to select rows and columns based on their labels or boolean conditions. It can be 

```python
# Select rows from index 4 to 8 (inclusive) from the DataFrame using label-based indexing
df.loc[4:8]
```

```python
# Filtering rows where the breed is between 'Poodle' or 'Husky' alphabetically
filtered_df = df.loc[(df['breed'] == 'Poodle') | (df['breed'] == 'Husky')]

# Displaying the filtered DataFrame
filtered_df
```

## ILoc

#### The .iloc[ ] function in Pandas is used for positional indexing, allowing you to select rows and columns based on their integer positions (i.e., row and column numbers). It works exclusively with numerical indices rather than labels.

```python
# iloc
df.iloc[4:7]
```
